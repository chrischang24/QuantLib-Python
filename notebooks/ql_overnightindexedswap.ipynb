```python
# define Common variable
class CommonVars:

    def __init__(self):
        self.today = ql.Date(5,2,2009)
        ql.Settings.instance().evaluationDate = self.today
        self.settlementDays = 2
        self.type = ql.OvernightIndexedSwap.Payer
        self.nominal = 100.0

        self.fixedEONIAPeriod = ql.Period(1, ql.Years)
        self.floatingEONIAPeriod = ql.Period(1, ql.Years)
        self.fixedEONIAConvention = ql.ModifiedFollowing
        self.floatingEONIAConvention = ql.ModifiedFollowing
        self.fixedEONIADayCount = ql.Actual360()
        self.eoniaTermStructure = ql.RelinkableYieldTermStructureHandle()
        self.EONIAIndex = ql.Eonia(self.eoniaTermStructure)
        self.calendar = self.EONIAIndex.fixingCalendar()

        self.fixedSwapFrequency = ql.Annual
        self.fixedSwapDayCount = ql.Thirty360(ql.Thirty360.ISDA)
        self.fixedSwapConvention = ql.ModifiedFollowing
        self.swapTermStructure = ql.RelinkableYieldTermStructureHandle()
        self.swapIndex = ql.Euribor3M(self.swapTermStructure)

        self.settlementDate = self.calendar.advance(self.today, ql.Period(self.settlementDays, ql.Days), ql.Following)

    def makeOISSwap(self, length, fixedRate, spread) -> ql.OvernightIndexedSwap:
        ois = ql.MakeOIS(length, self.EONIAIndex, fixedRate, 
                         overnightLegSpread=spread,
                         effectiveDate=self.settlementDate,
                         nominal=self.nominal,
                         discountingTermStructure=self.eoniaTermStructure)
        return ois
```

```python
# Test EONIA swap curve building
vars = CommonVars()

eoniaHelpers = []
swap3MHelpers = []

euribor3M = ql.Euribor3M()
eonia = ql.Eonia()

# initialize short end rate helpers
for i in range(len(depositData)):
    rate = 0.01 * depositData[i].rate
    simple = ql.SimpleQuote(rate)
    quote = ql.QuoteHandle(simple)
    term = ql.Period(depositData[i].n, depositData[i].timeUnit)
    helper = ql.DepositRateHelper(quote,
                                  term,
                                  depositData[i].settlementDays,
                                  euribor3M.fixingCalendar(),
                                  euribor3M.businessDayConvention(),
                                  euribor3M.endOfMonth(),
                                  euribor3M.dayCounter())
    
    if term <= ql.Period(2, ql.Days):
        eoniaHelpers.append(helper)
    
    if term <= ql.Period(3, ql.Months):
        swap3MHelpers.append(helper)

# initialize short end (fra part) rate helpers
for i in range(len(fraData)):
    rate = 0.01 * fraData[i].rate
    simple = ql.SimpleQuote(rate)
    quote = ql.QuoteHandle(simple)
    helper = ql.FraRateHelper(quote,
                            fraData[i].nExpiry,
                            fraData[i].nMaturity,
                            fraData[i].settlementDays,
                            euribor3M.fixingCalendar(),
                            euribor3M.businessDayConvention(),
                            euribor3M.endOfMonth(),
                            euribor3M.dayCounter())
    swap3MHelpers.append(helper)

# initialize long end - EONIA swap
for i in range(len(eoniaSwapData)):
    rate = 0.01 * eoniaSwapData[i].rate
    simple = ql.SimpleQuote(rate)
    quote = ql.QuoteHandle(simple)
    term = ql.Period(eoniaSwapData[i].n, eoniaSwapData[i].timeUnit)
    helper = ql.OISRateHelper(eoniaSwapData[i].settlementDays,
                              term,
                              quote,
                              eonia)
    eoniaHelpers.append(helper)

# initialize long end - Euribor 3M swap
for i in range(len(swapData)):
    rate = 0.01 * swapData[i].rate
    simple = ql.SimpleQuote(rate)
    quote = ql.QuoteHandle(simple)
    tenor = ql.Period(swapData[i].nIndexUnits, swapData[i].indexUnit)
    term = ql.Period(swapData[i].nTermUnits, swapData[i].termUnit)
    helper = ql.SwapRateHelper(quote,
                               term,
                               vars.calendar,
                               vars.fixedSwapFrequency,
                               vars.fixedSwapConvention,
                               vars.fixedSwapDayCount,
                               euribor3M)
    if tenor == ql.Period(3, ql.Months):
        swap3MHelpers.append(helper)

eoniaTS = ql.PiecewiseLogLinearDiscount(vars.today,
                                        eoniaHelpers,
                                        ql.Actual365Fixed())
swapTS = ql.PiecewiseLogLinearDiscount(vars.today,
                                       swap3MHelpers,
                                       ql.Actual365Fixed())

vars.eoniaTermStructure.linkTo(eoniaTS)

# check curve consistency
tolerance = 1.0e-10
for i in range(len(eoniaSwapData)):
    expected = eoniaSwapData[i].rate
    term = ql.Period(eoniaSwapData[i].n, eoniaSwapData[i].timeUnit)
    ois = vars.makeOISSwap(term, 0.0, 0.0)
    calculated = ois.fairRate() * 100.0
    
    print(f'term {term}, calculated rate: {calculated}')
    
    if abs(expected-calculated) > tolerance:
        print(f"curve inconsistency: ois swap length {term}, quoted rate: {expected}, claculated rate: {calculated}")

```
